```python
import pandas as pd
from datetime import datetime
import streamlit as st
import joblib

modelo = joblib.load('modelo.joblib')
```

```python
df = pd.read_csv('base_predict.csv')
df['horario'] = pd.to_datetime(df['horario'], format='%H:%M').apply(lambda x: x.strftime('%H:%M'))
df = df[df['horario'] >= '15:30']
df
```

```python
time_casa = []
player_casa = []
player_fora = []
time_fora = []

for i in df.itertuples():
    time_casa.append('time_casa' + '_' + i[3])
    time_fora.append('time_fora' + '_' + i[8])
    player_casa.append('player_casa' + '_' + i[4])
    player_fora.append('player_fora' + '_' + i[7])

data = {}
data['time_casa'] = time_casa
data['player_casa'] = player_casa
data['player_fora'] = player_fora
data['time_fora'] = time_fora

valores = pd.DataFrame(data)
previsao = modelo.predict(valores)
    
```
