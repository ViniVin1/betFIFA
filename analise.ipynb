```python
import pandas as pd
import numpy as np
import tqdm
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression, ElasticNet
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('base_jogos.csv').drop(['data', '+0.5', '+1.5', '+2.5', '+3.5', 'vencedor'], axis=1)
df.head()
```

```python
dummies = ['time_casa', 'player_casa', 'player_fora', 'time_fora']
df_dummie = pd.get_dummies(data=df, columns=dummies)
df_dummie
```

```python
def avaliar_modelo(nome_modelo, y_teste, previsao):
    r2 = r2_score(y_teste, previsao)
    rsme = np.sqrt(mean_squared_error(y_teste, previsao))
    return f'Modelo {nome_modelo}:\nR2:{r2:.2%}\nRSME:{rsme:.2f}'
```

```python
model_rf = RandomForestRegressor()
model_lr = LinearRegression()
model_et = ExtraTreesRegressor()

model_gb = GradientBoostingRegressor()
model_en = ElasticNet()
model_svr = SVR()

# modelos = {
#     'RandomForest': model_rf,
#     'LinearRegression': model_lr,
#     'ExtraTrees': model_et
# }

modelos = {
    'Gradient Boosting Regressor': model_gb,
    'Elastic Net': model_en,
    'Support Vector Regression': model_svr
}


# y = df_dummie[['gols_casa', 'gols_fora', 'total_gols']]
y = df_dummie['total_gols']
x = df_dummie.drop(['gols_casa', 'gols_fora', 'total_gols'], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=10)

for nome_modelo, modelo in modelos.items():
    #Treino
    modelo.fit(x_train, y_train)

    #Teste
    previsao = modelo.predict(x_test)
    print(avaliar_modelo(nome_modelo, y_test, previsao))
```
