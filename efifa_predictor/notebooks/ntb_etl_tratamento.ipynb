```python
import pandas as pd
import datetime
from datetime import date as dt
```

```python
colunas_new = {
    'data': 'data_hora',
    'team_player_1': 'time_casa',
    'player_1': 'jogador_casa',
    'score_player_1': 'gols_casa',
    'score_player_2': 'gols_fora',
    'player_2': 'jogador_fora',
    'team_player_2': 'time_fora'
}
base_raw = pd.read_csv('../data/raw/raw_base_jogos.csv').rename(columns=colunas_new)
base_raw.head()
```

```python
base_raw['data'] = pd.to_datetime(base_raw['data_hora'].str.split('T').str[0]).dt.strftime("%d/%m/%Y")
base_raw['hora'] = pd.to_datetime(base_raw['data_hora']).dt.strftime('%H:%M')

display(base_raw)
```
