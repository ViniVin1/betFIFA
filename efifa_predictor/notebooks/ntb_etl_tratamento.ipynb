### ntb_etl_tratamento.ipynb

Notebook responsável por realizar o tratamento da base de jogos e exportar uma tabela confiável para treinamento do modelo de predição

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
base_raw = pd.read_csv('../data/raw/base_partidas_raw.csv')
base_raw.head()
```

#### Filtra as tabelas necessárias

```python
base_copy = base_raw.copy() # Cria cópia da base

base_filtered = base_copy[
    [
        'id', 
        'date',
        'tournament.id', 
        'tournament.token_international', 
        'console.id', 
        'console.token',
        'console.token_international',
        'participant1.id',
        'participant1.nickname',
        'participant1.score',
        'participant1.team.id',
        'participant1.team.token_international',
        'participant2.id',
        'participant2.nickname',
        'participant2.score',
        'participant2.team.id',
        'participant2.team.token_international'
    ]
]

```

#### Renomeia o estádios que vieram com nomes errados

```python
rename_estadios = {            # Renomeando o nome dos estádios que faltaram
    '1': 'Old Trafford',
    '2': 'Old Trafford',
    '3': 'Wembley',
    '4': 'Wembley',
    'console-12': 'Stamford Bridge',
    'console-11': 'Stamford Bridge'
}

base_filtered["console.token_international"] = (
    base_filtered["console.token_international"].replace(rename_estadios)
)
```

```python
# estadios = base_filtered[['console.id','console.token', 'console.token_international']].drop_duplicates()
# estadios
```

#### Cria as colunas de Data e Hora

```python
# Cria coluna Data
base_filtered.loc[:, 'data'] = (
    pd.to_datetime(base_filtered['date'].str.split('T').str[0], errors='coerce')
      .dt.strftime('%d/%m/%Y')
)

# Cria coluna Hora
hora_brt_dt = pd.to_datetime(base_filtered['date'], errors='coerce').dt.floor('min')

# Cria coluna hora_uk (3hrs atrás da brt)
hora_uk_dt = hora_brt_dt - pd.Timedelta(hours=3)

# 5) Format
base_filtered.loc[:, 'hora_brt'] = hora_brt_dt.dt.strftime('%H:%M')
base_filtered.loc[:, 'hora_us']  = hora_uk_dt.dt.strftime('%H:%M')
```

```python
base_filtered.head()
```

#### Renomeia as colunas da tabela

```python
renamed_columns = {

    'date': 'datetime',
    'tournament.id': 'id_torneio',
    'tournament.token_international': 'nome_torneio',
    'console.id': 'id_estadio',
    'console.token': 'token_estadio',
    'console.token_international': 'nome_estadio',
    'participant1.id': 'id_player_casa',
    'participant1.nickname': 'nome_player_casa',
    'participant1.score': 'gols_casa',
    'participant1.team.id': 'id_time_casa',
    'participant1.team.token_international': 'nome_time_casa',
    'participant2.id': 'id_player_fora',
    'participant2.nickname': 'nome_player_fora',
    'participant2.score': 'gols_fora',
    'participant2.team.id': 'id_time_fora',
    'participant2.team.token_international': 'nome_time_fora'

}

base_renamed = base_filtered.rename(columns=renamed_columns)
```

#### Calcula o total de gols e over x gols

```python
# Coluna Resultado
base_renamed['resultado'] = base_renamed.apply(
    lambda row: 'casa' if row['gols_casa'] > row['gols_fora']
    else 'fora' if row['gols_casa'] < row['gols_fora']
    else 'empate',
    axis=1
)

# Total de gols
base_renamed['gols_total'] = base_renamed['gols_casa'] + base_renamed['gols_fora']

# # Over Gols
# base_renamed['over_0_5'] = (base_renamed['gols_total'] > 0.5).astype(int)
# base_renamed['over_1_5'] = (base_renamed['gols_total'] > 1.5).astype(int)
# base_renamed['over_2_5'] = (base_renamed['gols_total'] > 2.5).astype(int)
# base_renamed['over_3_5'] = (base_renamed['gols_total'] > 3.5).astype(int)
```

#### ID resultado

```python
map_resultado = {'casa': 1, 'empate': 0, 'fora': 2}
base_renamed['id_resultado'] = base_renamed['resultado'].map(map_resultado)
```

#### Exporta a base tratada

```python
# Export base total
base_renamed.to_csv('../data/trusted/base_partidas_trusted.csv', index=False, encoding='utf-8')

# Export base sem features não numéricas
base_numerica = base_renamed.drop(['id', 'datetime', 'id_torneio', 'id_estadio', 'nome_torneio', 'token_estadio',
                                    'nome_estadio', 'nome_player_casa', 'nome_time_casa',
                                    'nome_player_fora', 'nome_time_fora', 'data', 'hora_brt',
                                    'hora_us', 'resultado'], axis=1)
base_numerica.to_csv('../data/trusted/base_numerica_trusted.csv', index=False, encoding='utf-8')
```
